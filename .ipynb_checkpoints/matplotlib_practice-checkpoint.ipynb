```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Method #1: Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Method #2: Create the plot with ax.plt(), Object-oriented interface method
fig, ax = plt.subplots()

# to define the axes
ax.plot(x_axis,y_axis)
```

```python
# Method #3: Create the plot with ax.plt(), another method of object-oriented
fig = plt.figure()
ax = fig.add_subplot()
ax. plot(x_axis,y_axis)
# To change figure-level attributes (such as axis labels, a title, or a legend) 
# or save the figure as an image, use fig = plt.figure().
```

```python
# Method #4: Quick grapg using object_oriented interface
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Using Object_oriented_interface method
#------------------------------------------------
# Change the line color to green.
# Change the marker to a diamond.
# Change the line width to 2 points.
ax = plt.axes()
ax.plot(x_axis, y_axis, marker="*", color="green", linewidth=2, label='Boston')
# Add a legend for the city of Boston.
ax.legend()
# Add a title and axes labels.
ax.set_xlabel("x label")
ax.set_ylabel("y label")
ax.set_title("title")
# Add a y-axis limit.
ax.set_ylim(0,45)
# Add grid lines.
ax.grid()
```

```python
# Create bar charts
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
plt.bar(x_axis,y_axis)
```

```python
# Create the plot.
plt.bar(x_axis,y_axis, color="green", label="Boston")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the horizontal plot
plt.barh(x_axis, y_axis)
```

```python
# Creat a horizontal bar with the ascending values of y.
plt.barh(x_axis,y_axis)
plt.gca().invert_yaxis()
```

```python
plt.barh(x_axis,y_axis, color="magenta", label="Boston")
plt.gca().invert_yaxis()
plt.xlabel("Date")
plt.ylabel("fare($)")
plt.title("PyBer Fare By Month")
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax=plt.subplots()
ax.bar(x_axis,y_axis)

```

```python
# Create a scatter plot
import matplotlib.pyplot as plt
x=[50, 42, 39, 34,32,48,22,17,8,13]
y=[17,19,21,22,24,23,34,31,43,35]
plt.scatter(x,y,c="green", marker="x", s=x, label="Chicago")
plt.xlabel("Total Number of Rides (per city)")
plt.xlim=(0,50)
plt.ylabel("Average Fare($)")
plt.title("Average Fare vs Total Rides")
plt.legend()
```

```python
# Create a bubble plot
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
plt.scatter(x_axis,y_axis, s=y_axis)
```

```python
# Make the y-values larger like bubbles
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis,y_axis, y_axis_larger)
  
```

```python
# Use comprehension list instead of for or while loop
# instead of using 
# y_axis_larger=[]
# for data in y_axis:
   # y_axis_larger.append(data*3)
# Use s=[i*3 for i in y_axis]
plt.scatter(x_axis,y_axis, s=[i*3 for i in y_axis])
```
